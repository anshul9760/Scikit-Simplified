# Dropping Null Values

```python
col=df.columns
print("Count of Null Values : ")
for i in range(len(col)):
    print(str(col[i])+" "+str(df[col[i]].isnull().sum()))
print("##############################################")
print("Now removing Null values less than 50")
for i in range(len(col)):
    if df[col[i]].isnull().sum() < 50:
        df = df[pd.notnull(df[col[i]])]
print("###############################################")
for i in range(len(col)):
    print(str(col[i])+" "+str(df[col[i]].isnull().sum()))
```

```python
fig = plt.figure(figsize = (15,5))
f1 = fig.add_subplot(1, 3, 1)
f1.set_ylim([0,100])
f2 = fig.add_subplot(1,3,2)
f2.set_ylim([0,100])
f3 = fig.add_subplot(1,3, 3)
f3.set_ylim([0,30])

df[col[0]].value_counts().head(10).plot(kind= "bar", title = "Top 10 Cities", ax = f1)
df[col[2]].value_counts().head(10).plot(kind= "bar", title = "Top 10  Domain", ax = f2)
df[col[6]].value_counts().head(10).plot(kind= "bar", title =  "Top 10  Fields", ax = f3)
plt.show()
```

# Merging Similar Values

```python
df.replace({'North Delhi': 'Delhi'}, inplace=True)
df.replace({'South Delhi': 'Delhi'}, inplace=True)
df.replace({'Startup Founder': 'Startup'}, inplace=True)
df.replace({'Startup Founder (Service-based)': 'Startup'}, inplace=True)
df.replace({'Startup Founder(Product-based)': 'Startup'}, inplace=True)
df.replace({'Startup Founder(Product-based),Startup Founder (Service-based)': 'Startup'}, inplace=True)
df.replace({'South Delhi': 'Delhi'}, inplace=True)
# A loop can also be performed
```

```python
print("Merging similar Values causes disimilarities in Graphs")
fig = plt.figure(figsize = (15,5))
f1 = fig.add_subplot(1, 3, 1)
f1.set_ylim([0,100])
f2 = fig.add_subplot(1,3,2)
f2.set_ylim([0,200])
f3 = fig.add_subplot(1,3, 3)
f3.set_ylim([0,30])

df[col[0]].value_counts().head(10).plot(kind= "bar", title = "Top 10 Cities", ax = f1)
df[col[2]].value_counts().head(10).plot(kind= "bar", title = "Top 10  Domain", ax = f2)
df[col[6]].value_counts().head(10).plot(kind= "bar", title =  "Top 10  Fields", ax = f3)
plt.show()
```
